# Using numpy for reading .CSV files

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
h_price_arr = np.loadtxt('/content/housePrice.csv', delimiter=',')
````

````python
h_price_arr = np.array(h_price)
h_price_arr[10, :]
````

````python
# a = np.array(['1', '2', '4'])
# a.astype(int)

h_price_arr[:, 0].astype(int)
````

````python
list(range(10, 20, 3))
````

````python
## Method 1
# H_PRICE = np.zeros(h_price_arr.shape)
# H_PRICE = np.zeros_like(h_price_arr)
# H_PRICE = h_price_arr
for i in range(h_price_arr.shape[0]):
    try:
        h_price_arr[i, 0] = int(h_price_arr[i, 0])
    except:
        h_price_arr[i, 0] = np.nan

print(h_price_arr)
````

````python
## Method 2
H_PRICE = np.empty((0, 8))
for i in range(h_price_arr.shape[0]):
    temp = h_price_arr[i, :]
    try:
        temp[0] = int(h_price_arr[i, 0])
        H_PRICE = np.vstack((H_PRICE, temp))
    except:
        continue

H_PRICE.shape
````

````python
plt.figure(figsize=(15, 6))

plt.subplot(131), plt.hist(h_price_arr[:, 0], bins=50)  # Area
plt.subplot(132), plt.hist(h_price_arr[:, 1], bins=50)  # Room
plt.subplot(133), plt.hist(h_price_arr[:, -1], bins=50)  # Price(USD)

plt.show()
````

````python
# plt.scatter(h_price_arr[:, 0], h_price_arr[:, -1])
plt.figure(figsize=(15, 6))
plt.subplot(121)
plt.plot(h_price_arr[:, 0], h_price_arr[:, -1], '.k', alpha=0.5)
plt.subplot(122)
plt.plot(h_price_arr[:, 0], h_price_arr[:, -1], '.k', alpha=0.3)
plt.xlim([np.min(h_price_arr[:, 0]), 250])
plt.ylim([np.min(h_price_arr[:, -1]), 1e6])
plt.show()
````

````python
# h_price_arr = np.nan_to_num(h_price_arr, 0)

colors = ['b', 'r']

plt.figure(figsize=(8, 8))
for i in range(1000):
    # plt.scatter(h_price_arr[i, 0], h_price_arr[i, -1], alpha=0.5, s=h_price_arr[i, 1]**4, c='k', marker='o')
    plt.scatter(h_price_arr[i, 0], h_price_arr[i, -1], alpha=0.5,
                s=h_price_arr[i, 1]**4, c=colors[int(h_price_arr[i, 4])], marker='o')
plt.xlim([np.min(h_price_arr[:, 0]), 250])
plt.ylim([np.min(h_price_arr[:, -1]), 1e6])

plt.grid(True)

plt.title('House price in USD vs. Area vs. #Rooms vs. Elevator')
plt.xlabel('Area $m^2$')
plt.ylabel('Price $')

plt.show()
````

````python
PaviaU[:10, :10, 0]
````

````python
plt.figure(figsize=(13, 13))

plt.subplot(121), plt.imshow(PaviaU[:, :, 100], cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(PaviaU_gt[:, :], cmap='jet')
plt.xticks([]), plt.yticks([])

plt.show()
````

````python
np.unique(PaviaU_gt)
````

````python
plt.figure(figsize=(10, 15))
plt.tight_layout()
for i in range(1, 10): 
    plt.subplot(3, 3, i)
    plt.xticks([]), plt.yticks([])
    plt.imshow(PaviaU_gt == i)
plt.show()
````

````python
plt.figure(figsize=(13, 13))

plt.subplot(121), plt.imshow(PaviaU[-200:, :150, 100], cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(PaviaU_gt[-200:, :150], cmap='jet')
plt.xticks([]), plt.yticks([])

plt.show()
````

````python
plt.figure(figsize=(13, 13))

crop = PaviaU[-200:, :150, 100].copy()
crop[PaviaU_gt[-200:, :150] != 8] = 0 
plt.subplot(121), plt.imshow(crop, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(PaviaU_gt[-200:, :150] == 8, cmap='jet')
plt.xticks([]), plt.yticks([])

plt.show()
````

````python
spectrum = PaviaU[-200:, :150, :][PaviaU_gt[-200:, :150] == 2]
spectrum.shape
````

````python
plt.figure(figsize=(15, 15))
plt.tight_layout()

for i in range(1, 10):
    plt.subplot(3, 3, i)
    spectrum = PaviaU[:, :, :][PaviaU_gt[:, :] == i]
    # plt.plot(x, y)
    # plt.plot([300, 310, 320, 330, ..., 2500], spectrum.T)
    plt.plot(spectrum.T, 'k', alpha=0.05)
    plt.ylim([0, 7000])

    if i % 3 != 1:
        plt.yticks([])

plt.savefig('/content/spectrum.png', dpi=300)
plt.savefig('/content/spectrum.svg')
plt.show()
````

# How to create a Conda Environment os your OS

In order to create:

```
conda create --prefix="D:\path\to\destination" --no-default-packages python=3.7 numpy
```

In order to activate:

```
activate "D:\path\to\destination"

pip install matplotlib
```

To turn off:

```
deactivate
```
