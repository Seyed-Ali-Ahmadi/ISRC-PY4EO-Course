```python
import numpy as np
# import matplotlib.pyplot as plt
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 10, 40)
x
```

```python
sinx = np.sin(x)
```

```python
plt.plot(x, sinx)
```

```python
cosx = np.cos(x)
plt.plot(x, cosx, 'x')
plt.plot(x, sinx, 'o', color = 'k')
plt.legend(['ali', 'gholi'])
```

```python
plt.figure()
plt.subplot(121)
plt.plot(x,sinx, color = 'b'), plt.xlabel('numbers'), plt.ylabel('outs')
plt.title('sine of x')
plt.subplot(122)
plt.plot(x, cosx, 'o', color = 'g')
plt.xlabel('numbers'), plt.ylabel('outs'), plt.title('cosine of x')
```

```python
np.inner(sinx, cosx)
```

```python
a = np.outer(sinx, cosx)
a.shape
```

```python
plt.imshow(a, cmap = 'hot')
```
